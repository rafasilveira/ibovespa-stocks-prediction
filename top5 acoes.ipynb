```python
# Inicializacoes
import pandas as pd
import numpy as np
import plotly.express as px

url_dataset='https://github.com/rafasilveira/ibovespa-stocks-prediction/blob/master/data/b3_stocks_1994_2020.csv?raw=true'
ticker='IVVB11'
```

```python
# carga de dados para uma acao
dataset = pd.read_csv(url_dataset)
dataset.info()
```

```python
# analise acoes mais populares em 2019
dataset['datetime'] = pd.to_datetime(dataset['datetime'])

clean_dataset = dataset.copy()
clean_dataset.drop(columns=['open', 'close', 'high', 'low'], inplace=True)
clean_dataset = clean_dataset[clean_dataset['datetime'].dt.year == 2019]
volume = clean_dataset.groupby(['ticker']).sum()
volume.sort_values('volume', ascending=False, inplace=True)
volume.head(5)
```

```python
# grafico
def get_close(ticker):
  data = dataset.copy()
  data = data[data['datetime'].dt.year > 2010]
  data = data[data['ticker'] == ticker]
  data.drop(columns=['open', 'high', 'low'], inplace=True)
  return data

petr4 = get_close('PETR4')
vale3 = get_close('VALE3')
itub4 = get_close('ITUB4')
bbdc4 = get_close('BBDC4')
bbas3 = get_close('BBAS3')

frames = [petr4, vale3, itub4, bbdc4, bbas3]

result = pd.concat(frames, ignore_index=False)

fig = px.line(result, x="datetime", y="close", color='ticker', title=f'Fechamento Ãºltimos 10 anos')
# fig = px.line(result, x="datetime", y="volume", color='ticker', title=f'Volume Top5')
fig.show()
```
